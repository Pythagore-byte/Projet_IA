```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"

from google.colab import drive
# drive.mount('/content/drive')
file_path = '/content/drive/My Drive/cleaned_ov_with_embeddings_filtered.csv'
df = pd.read_csv(file_path)

# Charger les données
# df = pd.read_csv('cleaned_ov_with_embeddings_filtered.csv')

# Préparer les champs numériques et les embeddings
# Convertir les embeddings en listes de nombres
df['embeddings'] = df['embeddings'].apply(lambda x: [float(i) for i in x.strip("[]").split(",")])

# Extraire et normaliser les caractéristiques
# Assurez-vous que chaque champ soit au format numérique
#df['collection_id'] = df['collection_id'].fillna(0).astype(float)
#df['genre_id'] = df['genre_id'].fillna(0).astype(float)
df['release_year_normalized'] = df['release_year_normalized'].astype(float)
df['runtime_normalized'] = df['runtime_normalized'].astype(float)

# Combiner toutes les caractéristiques pour former un seul vecteur d'entrée par film
features = ['release_year_normalized', 'runtime_normalized']
metadata = df[features].values
embeddings = np.array(df['embeddings'].tolist())
full_data = np.hstack((metadata, embeddings))  # Concatenation des données metadata et embeddings

# Paramètres pour le modèle
input_dim = full_data.shape[1]  # Dimension d'entrée
output_dim = 10  # Dimension compressée souhaitée

# Définition de UNet en 1D
class UNet1D(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(UNet1D, self).__init__()

        # Encoder
        self.enc1 = nn.Conv1d(1, 16, kernel_size=3, padding=1)
        self.enc2 = nn.Conv1d(16, 32, kernel_size=3, padding=1)
        self.enc3 = nn.Conv1d(32, 64, kernel_size=3, padding=1)

        # Bottleneck
        self.bottleneck = nn.Conv1d(64, 128, kernel_size=3, padding=1)

        # Decoder
        self.dec3 = nn.ConvTranspose1d(128, 64, kernel_size=3, padding=1)
        self.dec2 = nn.ConvTranspose1d(64, 32, kernel_size=3, padding=1)
        self.dec1 = nn.ConvTranspose1d(32, 16, kernel_size=3, padding=1)

        # Output layer
        self.output_layer = nn.Conv1d(16, 1, kernel_size=3, padding=1)

        # Compression layer to get the compressed embedding
        self.compress_layer = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        # Encoder
        x = torch.relu(self.enc1(x))
        x = torch.relu(self.enc2(x))
        x = torch.relu(self.enc3(x))

        # Bottleneck
        x = torch.relu(self.bottleneck(x))

        # Decoder
        x = torch.relu(self.dec3(x))
        x = torch.relu(self.dec2(x))
        x = torch.relu(self.dec1(x))

        # Output
        x = self.output_layer(x)

        # Flatten and compress to output_dim
        x_flat = x.view(x.size(0), -1)  # Flatten
        compressed_output = self.compress_layer(x_flat)

        return compressed_output

# Convertir les données en tenseurs
full_data_tensor = torch.tensor(full_data, dtype=torch.float32).unsqueeze(1).to(device)  # Ajouter une dimension pour Conv1d

# Initialiser le modèle et optimiser
model = UNet1D(input_dim=input_dim, output_dim=output_dim).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

# Entraînement simplifié
epochs = 10000
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(full_data_tensor)
    loss = loss_fn(output, full_data_tensor.view(full_data_tensor.size(0), -1)[:, :output_dim])  # Calculer la perte pour la dimension compressée
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

# Obtenir les vecteurs compressés
with torch.no_grad():
    compressed_embeddings = model(full_data_tensor).detach().cpu().numpy()

# Ajouter les embeddings compressés au DataFrame et sauvegarder
df['compressed_embedding'] = compressed_embeddings.tolist()
output_file_path = '/content/drive/My Drive/main_with_compressed_embeddings_unet.csv'
df.to_csv(output_file_path, index=False)
print(f"Les vecteurs compressés ont été générés et sauvegardés dans {output_file_path}.")

```
